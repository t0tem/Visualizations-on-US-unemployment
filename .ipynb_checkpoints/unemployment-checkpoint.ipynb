```python
import pandas as pd
import numpy as np
unrate = pd.read_csv('unrate.csv')
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate.head(12)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(unrate.loc[:11, 'DATE'],
         unrate.loc[:11, 'VALUE'])
plt.show()
```

## X-labels rotated

```python
data = unrate[0:12]
plt.plot(data['DATE'],
         data['VALUE'])
plt.xticks(rotation=90)
plt.show()
```

## With labels

```python
data = unrate[0:12]
plt.plot(data['DATE'],
         data['VALUE'])
plt.xticks(np.arange(12), rotation=90)
plt.xlabel('Month')
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
plt.show()
```
